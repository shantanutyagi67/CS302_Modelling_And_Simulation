```python
import os
import numpy as np
from matplotlib import pyplot as plt
import numpy
import scipy
import pandas as pd
import random
```

```python
# position v/s time with varying runs
# unbiased
runs = [1,1,1]
seed = 22
steps = 50
random.seed(seed)
for r in runs:
    pos = np.zeros(steps)
    for j in range(len(pos)):
        val = np.random.uniform(0,1)
        if(val<0.5):
            val=-1
        else:
            val =1
        pos[j] = pos[j-1] + val
        
    plt.grid(b = True, color ='grey',  linestyle ='-.', linewidth = 0.5, alpha = 0.6)
    plt.xlabel('Steps')
    plt.ylabel('Position')
    plt.plot(pos)
plt.title('Unbiased walk ')
plt.show()
```

```python
# position v/s time with varying runs
# unbiased
runs = [10, 100,1000]
seed = 22
steps = 50
random.seed(seed)
for r in runs:
    net_pos = []
    for i in range(1,r):
        pos = np.zeros(steps)
        for j in range(len(pos)):
            val = np.random.uniform(0,1)
            if(val<0.5):
                val=-1
            else:
                val =1
            pos[j] = pos[j-1] + val
        net_pos.append(pos)
    net_pos = np.array(net_pos)
    pos = []
    # pos = pos/r
    for i in range(steps):
        pos.append(np.average(net_pos[:,i]))
        
    plt.grid(b = True, color ='grey',  linestyle ='-.', linewidth = 0.5, alpha = 0.6)
    plt.xlabel('Steps')
    plt.ylabel('Average Position')
    plt.plot(pos)
plt.title('Unbiased walk ')
plt.legend(['10 runs','100 runs','1000 runs'])
plt.show()
```

```python
# variance v/s time with varying runs
# unbiased
runs = [10,100,1000]
seed = 22
steps =50
random.seed(seed)
for r in runs:
    net_pos = []
    var = []
    for i in range(r):
        pos = np.zeros(steps)
        for j in range(len(pos)):
            val = np.random.uniform(0,1)
            if(val<0.5):
                val=-1
            else:
                val =1
            pos[j] = pos[j-1] + val
        net_pos.append(pos)
    net_pos = np.array(net_pos)
    for i in range(steps):
        std = np.std(net_pos[:,i])
        var.append(std**2)
    plt.grid(b = True, color ='grey',  linestyle ='-.', linewidth = 0.5, alpha = 0.6)
    plt.xlabel('Steps')
    plt.ylabel('Variance')
    plt.plot(var)
plt.legend(['10 runs','100 runs','1000 runs'])
plt.title('Variance of Unbiased walk ')
plt.show()
```

```python
# q3 unbiased with 1000 runs
runs = 10000
seed = 22
steps = [1000, 1100]
random.seed(seed)
c= ['cyan','magenta']
ind = 0
for s in steps:
    net_pos = []
    for i in range(runs):
        pos = np.zeros(s)
        for j in range(len(pos)):
            val = np.random.uniform(0,1)
            if(val<0.5):
                val=-1
            else:
                val =1
            pos[j] = pos[j-1] +  val
        net_pos.append(pos)
    net_pos = np.array(net_pos)
    net_pos = net_pos.flatten()

    plt.grid(b = True, color ='grey',  linestyle ='-.', linewidth = 0.5, alpha = 0.5)
    plt.xlabel('Distance')
    plt.ylabel('Probability')
    plt.hist(net_pos,density=True,bins=20,alpha=0.7,color=c[ind])
    ind +=1
    plt.title(' Histogram for Probability Distribution')
plt.legend(['1000 steps','1100 steps'])
plt.show()
```
