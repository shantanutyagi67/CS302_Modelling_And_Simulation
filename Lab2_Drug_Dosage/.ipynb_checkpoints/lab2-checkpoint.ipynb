```python
import numpy as np
from matplotlib import pyplot as plt
import math
```

```python
# Exercise q5 with single compartment
# 1 compartment model for Drug Dosage with Dirac Delta as administration time
# Dilantin
initial_val=33.33
dosage=100
t_end=24*7
mec=10 
mtc=20
step_size=0.001
half_life=22
interval = 24
rb=math.log(2)/half_life
absorbing_fraction=0.12
total_steps= int(t_end/step_size)+1 
quantity_a = np.zeros(total_steps)
quantity_a[0]=absorbing_fraction* initial_val
check=0
for i in range(1,total_steps):
    if(((i)*step_size)%interval ==0):
        quantity_a[i-1]+=absorbing_fraction* dosage
      
    quantity_a[i] = quantity_a[i-1]-rb*quantity_a[i-1]*step_size
    if(quantity_a[i]>=mtc and check==0):
        print("MTC exceeded at" + str(i*step_size)+" days")
        check=1

x_axis=np.linspace(0,t_end,total_steps)

plt.figure(1)
plt.plot(x_axis,quantity_a,'r' )
plt.legend(['Concentration Of Medicine'])
plt.title('Drug Dosage')
plt.ylabel('Concentration (g/mL)')
plt.xlabel('Time in hours')
plt.grid(b = True, color ='grey', linestyle ='-.', linewidth = 0.5,
        alpha = 0.6)
plt.axhline(mec,color='y')
plt.axhline(mtc,color='m')
plt.legend(['Conc in Compartment 1','Minimum Effective Concentration','Minimum Toxic Concentration'])
plt.show()
```

```python
# Exercise q5 with single compartment
# 1 compartment model for Drug Dosage with step function as administration time
# Dilantin
initial_val=33.33
dosage=100
t_end=24*7
mec=10
mtc=20
step_size=0.001
half_life=22
interval = 24
rb=math.log(2)/half_life
absorbing_fraction=0.12
total_steps= int(t_end/step_size)+1 
quantity_a = np.zeros(total_steps)
quantity_a[0]=absorbing_fraction* initial_val
admTime = 0.3333
# counter=np.linspace(0,int(admTime/step_size),step_size)
start = False
counter=admTime/step_size
check=0
j=0
for i in range(1,total_steps):
    if(((i)*step_size)%interval ==0):
        start = True
    if start == True and j>=counter :
        j=0
        start= False
    if start==True:
        quantity_a[i-1]+=absorbing_fraction* dosage/ counter
        j+=1
      
    quantity_a[i] = quantity_a[i-1]-rb*quantity_a[i-1]*step_size
    if(quantity_a[i]>=mtc and check==0):
        print("MTC exceeded at" + str(i*step_size)+" days")
        check=1


x_axis=np.linspace(0,t_end,total_steps)

plt.figure(1)
plt.plot(x_axis,quantity_a,'r' )
plt.legend(['Concentration Of Medicine'])
plt.title('Drug Dosage')
plt.ylabel('Concentration (g/mL)')
plt.xlabel('Time in hours')
plt.grid(b = True, color ='grey', linestyle ='-.', linewidth = 0.5,
        alpha = 0.6)
plt.axhline(mec,color='y')
plt.axhline(mtc,color='m')
plt.legend(['Conc in Compartment 1','Minimum Effective Concentration','Minimum Toxic Concentration'])
plt.show()
```

```python

# 1 compartment model with step function
# Dilantin
dosage=33.333
t_end=24*5
mec=10
mtc=20
step_size=0.001
half_life=22
interval = 8
rb=math.log(2)/half_life
absorbing_fraction=0.12
total_steps= int(t_end/step_size)+1 
quantity_a = np.zeros(total_steps)
quantity_a[0]=absorbing_fraction* dosage
admTime = 0.3333
# counter=np.linspace(0,int(admTime/step_size),step_size)
start = False
counter=admTime/step_size

j=0
for i in range(1,total_steps):
    if(((i)*step_size)%interval ==0):
        start = True
    if start == True and j>=counter :
        j=0
        start= False
    if start==True:
        quantity_a[i-1]+=absorbing_fraction* dosage/ counter
        j+=1
      
    quantity_a[i] = quantity_a[i-1]-rb*quantity_a[i-1]*step_size

x_axis=np.linspace(0,t_end,total_steps)

plt.figure(1)
plt.plot(x_axis,quantity_a,'r' )
plt.legend(['Concentration Of Medicine'])
plt.title('Drug Dosage')
plt.ylabel('Concentration (g/mL)')
plt.xlabel('Time in hours')
plt.grid(b = True, color ='grey', linestyle ='-.', linewidth = 0.5,
        alpha = 0.6)
plt.axhline(mec,color='y')
plt.axhline(mtc,color='m')
plt.legend(['Conc in Compartment 1','Minimum Effective Concentration','Minimum Toxic Concentration'])
plt.show()
# print(quantity_a[int(48/step_size)])
```

```python
# 2 compartment model of aspirin with 1 dosage
#  ra=7rb
initial_val=2*325*1000/3000
t_end=24*1
mec=150
mtc=350
step_size=0.001
half_life=3.2
interval = 8
rb=math.log(2)/half_life
ra=7*rb
total_steps= int(t_end/step_size)+1 
quantity_a = np.zeros(total_steps)
quantity_b = np.zeros(total_steps)
# absorbtion frac is only for comp to blood stream
quantity_a[0]=initial_val
for i in range(1,total_steps):
    
    quantity_a[i] = quantity_a[i-1]-ra*quantity_a[i-1]*step_size
    quantity_b[i] = quantity_b[i-1] + (ra*quantity_a[i-1] - rb*quantity_b[i-1])*step_size

x_axis=np.linspace(0,t_end,total_steps)

plt.figure(1)
# plt.plot(x_axis,quantity_a,'r' )
plt.plot(x_axis,quantity_a,'r-' )
plt.plot(x_axis,quantity_b,'g-')

plt.title('Drug Concentration variation in Bloodstream')
plt.ylabel('Concentration (g/mL)')
plt.xlabel('Time iin hours')
plt.grid(b = True, color ='grey', linestyle ='-.', linewidth = 0.5,
        alpha = 0.6)
plt.axhline(mec,color='y')
plt.axhline(mtc,color='m')
plt.legend(['Drug Concentration in GI tract','Conc in Bloodstream','Minimum Effective Concentration','Minimum Toxic Concentration'],bbox_to_anchor=(1.05, 1), loc=2)
# plt.legend()
plt.show()


```

```python
# 2 compartment model with multiple dosage (dirac delta)
# generic dilantin
# ra=2rb
initial_val=33.33
t_end=24*6
mec=10
mtc=20
step_size=0.001
half_life=22
interval = 8
rb=math.log(2)/half_life
ra=2*rb
absorbing_fraction=0.12
total_steps= int(t_end/step_size)+1 
quantity_a = np.zeros(total_steps)
quantity_b = np.zeros(total_steps)
# absorbtion frac is only for comp to blood stream
quantity_a[0]=initial_val
dosage=initial_val
for i in range(1,total_steps):
    if(((i)*step_size)%interval == 0):
        quantity_a[i-1]+=dosage
    quantity_a[i] = quantity_a[i-1]-ra*quantity_a[i-1]*step_size
    quantity_b[i] = quantity_b[i-1] + (ra*quantity_a[i-1]*absorbing_fraction - rb*quantity_b[i-1])*step_size

x_axis=np.linspace(0,t_end,total_steps)

plt.figure(1)
# plt.plot(x_axis,quantity_a,'r' )
plt.plot(x_axis,quantity_b,'g')
plt.title('Drug Concentration variation in Bloodstream')
plt.ylabel('Concentration (g/mL)')
plt.xlabel('Time in hours')
plt.grid(b = True, color ='grey', linestyle ='-.', linewidth = 0.5,
        alpha = 0.6)
plt.axhline(mec,color='y')
plt.axhline(mtc,color='m')
plt.legend(['Conc in Compartment 2','Minimum Effective Concentration','Minimum Toxic Concentration'],bbox_to_anchor=(1.05, 1), loc=2)
# plt.legend()
plt.show()
plt.plot(x_axis,quantity_a,'r' )
plt.grid(b = True, color ='grey', linestyle ='-.', linewidth = 0.5,
        alpha = 0.6)
plt.title('Drug Concentration variation in GI tract')
plt.ylabel('Concentration (g/mL)')
plt.xlabel('Time in hours')
plt.show()
```
