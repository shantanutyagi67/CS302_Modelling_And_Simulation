```python
import os
import numpy as np
from matplotlib import pyplot as plt
import numpy
import scipy
import pandas as pd

# air conditioner
# only external influence
initial_val=0
p = 0.010399
p1 = 0.02
p2 = 0.03
q = 0.41861
t_end=200
na = 1
step_size=0.1
total_steps= int(t_end/step_size)+1 
quantity_a = np.zeros(total_steps)
quantity_a1 = np.zeros(total_steps)
quantity_a2 = np.zeros(total_steps)
quantity_b = np.zeros(total_steps)
quantity_b1 = np.zeros(total_steps)
quantity_b2 = np.zeros(total_steps)

quantity_a[0]=initial_val
quantity_a1[0]=initial_val
quantity_a2[0]=initial_val
quantity_b[0]=initial_val
quantity_b1[0]=initial_val
quantity_b2[0]=initial_val

for i in range(1,total_steps):
    quantity_a[i] = quantity_a[i-1] + (p )*(na-quantity_a[i-1])*step_size
    quantity_a1[i] = quantity_a1[i-1] + (p1 )*(na-quantity_a1[i-1])*step_size
    quantity_a2[i] = quantity_a2[i-1] + (p2 )*(na-quantity_a2[i-1])*step_size
    
    quantity_b[i-1] = (p )*(na-quantity_a[i-1])
    quantity_b1[i-1] = (p1 )*(na-quantity_a1[i-1])
    quantity_b2[i-1] = (p2 )*(na-quantity_a2[i-1])
time=quantity_b.argmax(axis=0)
x_axis=np.linspace(0,1,total_steps)
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.plot(x_axis,quantity_a,'r' )
ax1.plot(x_axis,quantity_a1,'g' )
ax1.plot(x_axis,quantity_a2,'b' )

ax2.plot(x_axis,quantity_b,'r--' )
ax2.plot(x_axis,quantity_b1,'g--' )
ax2.plot(x_axis,quantity_b2,'b--' )


ax1.set_title('External influence Model')
ax1.set_ylabel('Fraction of Population')
ax2.set_ylabel('Rate of Adoption')
ax1.set_xlabel('Time')
plt.grid(b = True, color ='grey',  
        linestyle ='-.', linewidth = 0.5,  
        alpha = 0.6)
print(time)
ax2.legend(['p=0.01','p=0.02','p=0.03',], loc=2)
ax1.legend(['p=0.01','p=0.02','p=0.03',], loc=1)

plt.xlim(left=0)
plt.show()
```

```python
import os
import numpy as np
from matplotlib import pyplot as plt
import numpy
import scipy
import pandas as pd

initial_val=0
p = 0.010399
q = 0.41861
t_end=65
na = 1
step_size=0.1
total_steps= int(t_end/step_size)+1 
quantity_a = np.zeros(total_steps)
quantity_b = np.zeros(total_steps)
quantity_c = np.zeros(total_steps)
quantity_d = np.zeros(total_steps)

quantity_a1 = np.zeros(total_steps)
quantity_b1 = np.zeros(total_steps)
quantity_c1 = np.zeros(total_steps)
quantity_d1 = np.zeros(total_steps)

quantity_a[0]=initial_val
quantity_b[0]=initial_val
quantity_c[0]=initial_val
quantity_d[0]=initial_val

quantity_a1[0]=initial_val
quantity_b1[0]=initial_val
quantity_c1[0]=initial_val
quantity_d1[0]=initial_val

b1 = -1
b2 = 1
b3 = 2
b4 = 0
for i in range(1,total_steps):
    quantity_a[i] = quantity_a[i-1] + (p + q*pow(quantity_a[i-1],b1+1)/na)*(na-quantity_a[i-1])*step_size
    quantity_b[i] = quantity_b[i-1] + (p + q*pow(quantity_b[i-1],b2)*quantity_b[i-1]/na)*(na-quantity_b[i-1])*step_size
    quantity_c[i] = quantity_c[i-1] + (p + q*pow(quantity_c[i-1],b3)*quantity_c[i-1]/na)*(na-quantity_c[i-1])*step_size
    quantity_d[i] = quantity_d[i-1] + (p + q*pow(quantity_d[i-1],b4)*quantity_d[i-1]/na)*(na-quantity_d[i-1])*step_size
    
    quantity_a1[i-1] = (p + q*pow(quantity_a[i-1],b1+1)/na)*(na-quantity_a[i-1])
    quantity_b1[i-1] = (p + q*pow(quantity_b[i-1],b2)*quantity_b[i-1]/na)*(na-quantity_b[i-1])
    quantity_c1[i-1] = (p + q*pow(quantity_c[i-1],b3)*quantity_c[i-1]/na)*(na-quantity_c[i-1])
    quantity_d1[i-1] = (p + q*pow(quantity_d[i-1],b4)*quantity_d[i-1]/na)*(na-quantity_d[i-1])
    
quantity_a1[i-1]=quantity_a1[i-2]
quantity_b1[i-1]=quantity_b1[i-2]
quantity_c1[i-1]=quantity_c1[i-2]
quantity_d1[i-1]=quantity_d1[i-2]

time1=quantity_a1.argmax(axis=0)
time2=quantity_b1.argmax(axis=0)
time3=quantity_c1.argmax(axis=0)
time4=quantity_d1.argmax(axis=0)


x_axis=np.linspace(0,1,total_steps)
plt.figure(1)
plt.plot(x_axis,quantity_a,'navy' )
plt.plot(x_axis,quantity_b,'red' )
plt.plot(x_axis,quantity_c,'orange' )
plt.plot(x_axis,quantity_d,'m' )


plt.legend(['beta = -1','beta = 1','beta = 2','beta = 0'])
plt.title('Modified Bass Model')
plt.ylabel('Fraction of Population')
plt.xlabel('Time')
plt.grid(b = True, color ='grey',  
        linestyle ='-.', linewidth = 0.5,  
        alpha = 0.6)
# print(time)
plt.plot(x_axis[time1],quantity_a[time1],marker='o',color='blue')
plt.plot(x_axis[time2],quantity_b[time2],marker='o',color='blue')
plt.plot(x_axis[time3],quantity_c[time3],marker='o',color='blue')
plt.plot(x_axis[time4],quantity_d[time4],marker='o',color='blue')

# plt.axvline(time*step_size/t_end)
plt.figure(2)
plt.plot(x_axis,quantity_a1,'navy' )
plt.plot(x_axis,quantity_b1,'red' )
plt.plot(x_axis,quantity_c1,'orange' )
plt.plot(x_axis,quantity_d1,'m' )
plt.legend(['beta = -1','beta = 1','beta = 2','beta = 0'])
plt.title('Modified Bass Model')
plt.ylabel('Rate of Adoption')
plt.xlabel('Time')
plt.grid(b = True, color ='grey',  
        linestyle ='-.', linewidth = 0.5,  
        alpha = 0.6)
# print(time)
plt.plot(x_axis[time1],quantity_a1[time1],marker='o',color='blue')
plt.plot(x_axis[time2],quantity_b1[time2],marker='o',color='blue')
plt.plot(x_axis[time3],quantity_c1[time3],marker='o',color='blue')
plt.plot(x_axis[time4],quantity_d1[time4],marker='o',color='blue')

plt.show()
```
