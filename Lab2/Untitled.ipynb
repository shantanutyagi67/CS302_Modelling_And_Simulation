```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
initial_val=150
t_end=24*7
mec=10 
mtc=20
step_size=0.001
interval = 24
rb=[0.09902,0.031506,0.016503]
c = ['r','g','b']
absorbing_fraction=0.12
total_steps= int(t_end/step_size)+1 
quantity_a = np.zeros(total_steps)
for j in range(len(rb)):
    quantity_a[0]=absorbing_fraction* initial_val
    for i in range(1,total_steps):    
        quantity_a[i] = quantity_a[i-1]-rb[j]*quantity_a[i-1]*step_size

    x_axis=np.linspace(0,t_end,total_steps)
    plt.plot(x_axis,quantity_a,c[j] )
        
plt.legend(['Concentration Of Medicine'])
plt.title('Drug Dosage with varying elimination constants')
plt.ylabel('Concentration (μg/mL)')
plt.xlabel('Time in hours')
plt.grid(b = True, color ='grey', linestyle ='-.', linewidth = 0.5,
        alpha = 0.6)
plt.axhline(mec,color='y')
plt.axhline(mtc,color='m')
plt.legend(['a = 0.09902','a = 0.031506','a = 0.016503','MEC','MTC'])
plt.show()
```

```python

import numpy as np
from matplotlib import pyplot as plt
import math
# Exercise q5 with single compartment
# 1 compartment model for Drug Dosage with step function as administration time
# Dilantin
initial_val=33.33
dosage=100
t_end=24*3
mec=10
mtc=20
step_size=0.001
half_life=22
interval = 24
rb=math.log(2)/half_life
absorbing_fraction=0.12
total_steps= int(t_end/step_size)+1 
quantity_a = np.zeros(total_steps)
quantity_a[0]=absorbing_fraction* initial_val
admTime = 1
# counter=np.linspace(0,int(admTime/step_size),step_size)
start = False
counter=admTime/step_size
check=0
j=0
for i in range(1,total_steps):
    if(((i)*step_size)%interval ==0):
        start = True
    if start == True and j>=counter :
        j=0
        start= False
    if start==True:
        quantity_a[i-1]+=absorbing_fraction* dosage/ counter
        j+=1
      
    quantity_a[i] = quantity_a[i-1]-rb*quantity_a[i-1]*step_size
    if(quantity_a[i]>=mtc and check==0):
        print("MTC exceeded at" + str(i*step_size)+" days")
        check=1


x_axis=np.linspace(0,t_end,total_steps)

plt.figure(1)
plt.plot(x_axis,quantity_a,'r' )
plt.legend(['Concentration Of Medicine'])
plt.title('Drug Dosage')
plt.ylabel('Concentration (μg/mL)')
plt.xlabel('Time in hours')
plt.grid(b = True, color ='grey', linestyle ='-.', linewidth = 0.5,
        alpha = 0.6)
plt.axhline(mec,color='y')
plt.axhline(mtc,color='m')
plt.legend(['Conc in Compartment 1','Minimum Effective Concentration','Minimum Toxic Concentration'])
plt.show()
```
